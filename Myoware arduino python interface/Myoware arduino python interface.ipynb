```python
import time
import serial
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
ser = serial.Serial('COM9',9600,timeout=1)
val_arr = []
i=0
while i<=1000:
    i = i+1
    val = ser.readline().decode('ascii')
    if val != '':
        val =int(val)
#     val_int_2 = int.from_bytes(val, byteorder="big")
        val_arr.append(val)
    #print(val_int_2)
    
#     print(type(val))
    
#     time.sleep(0.01)
# val_int.pop(0)
print(val_arr,"\n")
```

```python
val_arr_mean = np.mean(val_arr)
print(val_arr_mean)
```
